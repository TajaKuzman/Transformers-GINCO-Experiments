```python
import os
import time
import json
import utils
import parse
import fasttext
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

```

```python
columns = ["suitable"] + [f"run_{i}" for i in range(15)]
df = pd.read_csv("backup_25.csv")#.loc[:, columns]
y_true, y_pred = [], []

for i in range(15):
    y_true.extend(df.suitable.values.tolist())
    y_pred.extend(df[f"run_{i}"].values.tolist())
```

```python
utils.plot_cm(y_true, y_pred, labels=[1,0], title="Cumulative CM, suitable=1, nonsuitable=0", 
save="images/25_cumulative_CM.png", include_metrics=False, figsize=(4,4))
```

```python
f1s = list()
micro_f1s = list()
macro_f1s = list()
for i in range(15):
    from sklearn.metrics import f1_score
    from sklearn.dummy import DummyClassifier
    y_true = df.suitable.values
    y_pred = df[f"run_{i}"].values
    f1s.append(f1_score(y_true, y_pred))
    micro_f1s.append(f1_score(y_true, y_pred, average="micro"))
    macro_f1s.append(f1_score(y_true, y_pred, average="macro"))
```

```python
print("As is: ", np.mean(f1s), np.std(f1s))
print("Micro: ", np.mean(micro_f1s), np.std(micro_f1s))
print("Macro: ", np.mean(macro_f1s), np.std(macro_f1s))
```

```python
from sklearn.metrics import f1_score
from sklearn.dummy import DummyClassifier
for average in ["micro", "macro"]:
    dummy_f1s = list()
    print(f"{average=}")
    for i in range(15):
        y_true = df.suitable.values
        dummy_clf = DummyClassifier(strategy="stratified")
        dummy_clf.fit(y_true, y_true)
        y_pred = dummy_clf.predict(y_true)
        dummy_f1s.append(f1_score(y_true, y_pred, average=average))
    print(f"\tDummy clf: strategy: 'stratified', {np.mean(dummy_f1s):0.3} +/- {np.std(dummy_f1s):0.3}")
    p = stats.mannwhitneyu(micro_f1s if average=="micro" else macro_f1s, dummy_f1s, alternative='greater')[1]
    print(f"\t\t p_value: {p:0.3}")

    
    dummy_f1s = list()
    for i in range(15):
        y_true = df.suitable.values
        dummy_clf = DummyClassifier(strategy="most_frequent")
        dummy_clf.fit(y_true, y_true)
        y_pred = dummy_clf.predict(y_true)
        dummy_f1s.append(f1_score(y_true, y_pred, average=average))
    print(f"\tDummy clf: strategy: 'most_frequent', {np.mean(dummy_f1s):0.3} +/- {np.std(dummy_f1s):0.3}")
    p = stats.mannwhitneyu(micro_f1s if average=="micro" else macro_f1s, dummy_f1s, alternative='greater')[1]
    print(f"\t\t p_value: {p:0.3}")
```

# Which categories get misclassified?

```python
results = list()
run = 0
for run in range(15):
    misclassified = df.suitable != df[f"run_{run}"]
    for i in df[misclassified].index:
        y_true = df.loc[i, "suitable"]
        y_pred = df.loc[i, f"run_{run}"]
        primary = df.loc[i, "primary"]

        results.append({
            "primary": primary,
            "y_true": y_true,
            "y_pred": y_pred
        })
```

```python
print(pd.DataFrame(data=results).groupby("primary").count().sort_values("y_true", ascending=False)["y_true"].to_markdown())
```

```python
df2 = pd.DataFrame(data=results).groupby("primary").count().sort_values("y_true", ascending=False)
df2["misclassified_counts"] = df2.y_true
df2.drop(columns=["y_true", "y_pred"], inplace=True)

df2["counts_in_test_split"] = [15*sum(df.primary == c) for c in df2.index]
df2["misclassification_ratio"] = df2.misclassified_counts / df2.counts_in_test_split

df2.sort_values("misclassification_ratio", ascending=False, inplace=True)
df2
```

```python
print(df2.to_markdown())
```

# Which do not get misclassified?

```python
results_not_misclassified = list()
run = 0
for run in range(15):
    not_misclassified = df.suitable == df[f"run_{run}"]
    for i in df[not_misclassified].index:
        y_true = df.loc[i, "suitable"]
        y_pred = df.loc[i, f"run_{run}"]
        primary = df.loc[i, "primary"]

        results_not_misclassified.append({
            "primary": primary,
            "y_true": y_true,
            "y_pred": y_pred
        })
```

```python
df3 = pd.DataFrame(data=results_not_misclassified).groupby("primary").count().sort_values("y_true", ascending=False)
df3["not_misclassified_counts"] = df3.y_true
df3.drop(columns=["y_true", "y_pred"], inplace=True)

df3["counts_in_test_split"] = [15*sum(df.primary == c) for c in df3.index]
df3["not_misclassification_ratio"] = df3.not_misclassified_counts / df3.counts_in_test_split

df3.sort_values("not_misclassification_ratio", ascending=True, inplace=True)
df3
```

```python
print(df3.to_markdown())
```
